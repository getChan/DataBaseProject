```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='ko', tz=360)
```

```python
kw_list=["그날처럼"]
pytrends.build_payload(kw_list, cat=0, timeframe='2018-01-01 2018-02-08', geo='KR', gprop='')
```

```python
df = pytrends.interest_over_time()
df
```
