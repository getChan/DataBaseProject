```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='ko', tz=360)
```

```python
kw_list=["좋니"]
pytrends.build_payload(kw_list, cat=0, timeframe='2017-06-22 2018-08-22', geo='KR', gprop='')
```

```python
df = pytrends.interest_over_time()
df
```

```python
df = pytrends.get_historical_interest(kw_list, year_start=2018, month_start=8, day_start=1, hour_start=0, year_end=2018, month_end=11, day_end=2, hour_end=0, cat=0, geo='KR', gprop='', sleep=60)
df
```

```python
from matplotlib import pyplot as plt

plt.plot(df['좋니'])
```
