```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='ko', tz=360)
```

```python
kw_list=["닐로"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python
df =pytrends.interest_over_time()
df
```

```python
df = pytrends.get_historical_interest(kw_list, year_start=2018, month_start=1, day_start=1, hour_start=0, year_end=2018, month_end=2, day_end=1, hour_end=0, cat=0, geo='KR', gprop='', sleep=0)
df
```

```python
from matplotlib import pyplot as plt

plt.plot(df['닐로'])
```
